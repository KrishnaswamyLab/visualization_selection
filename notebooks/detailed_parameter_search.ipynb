```python
import warnings
warnings.simplefilter("ignore")
import pickle
from blog_tools import data, interact
%matplotlib inline
```

On the left, you see a 3D projection of the Swiss Roll. Move the sliders below the plot to adjust the parameters for the visualization on the right. _If the points don't appear on your screen, try moving your mouse over the plot or double clicking on the visualization._

```python
results = pickle.load(open("../data/parameter_search_detailed.pickle", 'rb'))
dataset = data.swissroll()
tab_widget = interact.TabWidget()
for algorithm in results.keys():
    with tab_widget.new_tab(algorithm):
        interact.parameter_plot(algorithm, results, dataset.X_true, c=dataset.c)

tab_widget.display()
```
