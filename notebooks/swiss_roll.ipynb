```python
import sklearn.datasets
import sklearn.decomposition
import scprep
X, t = sklearn.datasets.make_swiss_roll(500)
X[:,1]*=1.2
scprep.plot.scatter2d(sklearn.decomposition.PCA(2).fit_transform(X), c=t)
```

```python
import phate
scprep.plot.scatter2d(phate.PHATE(2).fit_transform(X), c=t)
```

```python
import sklearn.manifold
scprep.plot.scatter2d(sklearn.manifold.TSNE(2).fit_transform(X), c=t)
```

```python
import umap
scprep.plot.scatter2d(umap.UMAP(2).fit_transform(X), c=t)
```

```python
scprep.plot.scatter2d(sklearn.manifold.Isomap(2).fit_transform(X), c=t)
```
